# Week 2 vs New York Jets

"It's not whether you get knocked down, it's whether you get back up." - Vince Lombardi

This quote from Coach Lombardi is perfect after week 1's loss to Chicago. The Titans still have their whole season in front of them, and they have the talent and ability to correct their mistakes. Now, they must get back up after week 1, and prepare for their home opener against the Jets.

# Overall Preview

There's good news and bad news, we'll start with the bad news. The bad news is that the offense could not get much going outside of their two touchdown drives combining for 56 yards on 10 non-half/touchdown ending drives. They finished 31st in EPA/pass and 28th in EPA/rush. Levis had the 15th highest play action percent of his attempts while also having the 13th highest time-to-throw and 14th highest average depth of target. I think the Titans should use more play action because Levis had about 32% of his yards last season come from play action which ranked 5th. This will also hold the linebackers and allow his receivers to get open. 

Fortunately for the Titans, the Jets defense was not great, albeit against one of the better offenses in the league. The Jets ranked 23rd in EPA/pass and 30th in EPA/rush. While these rankings are far from what we saw last year, they were able to pressure the 49ers on about 24% of dropbacks which ranked 8th while only blitzing on 24% of dropbacks which was the 15th highest. The Jets blitzed more than the Bears (15%) which is another reason why I think play action will be good for Levis. The blitzers must take into account the running back on these play action passes which will slow down the pass rush.

The good news for the Titans is that their defense ranked 7th in EPA/pass as well as EPA/rush. They were able to get pressures on 21% of drop backs which ranked 13th and blitzed on 24% of dropbacks which ranked 15th. Getting pressure will be key against a quarterback coming off an achilles injury as a 40-year old.  One way for the Jets to avoid the Titans pressure would be to continue to get the ball out quick. Aaron Rodgers had an average time-to-throw of 2.1 seconds which was the 2nd quickest and led to a pressure rate of only 4.5% which was 5th best. While Rodgers was getting the ball out quick, he was still able to maintain the 11th highest average depth of target. The Jets will need this high depth of target to generate more explosive plays this week because they only generated 2 explosive pass plays which ranked 21st and 1 explosive run which ranked 27th. Finally, this will be a fun matchup to see if the Titans can keep the Jets off schedule because the Titans forced the Bears into the longest distance to go on 3rd downs while the Jets had the second fewest yards to go on average last week.

# Trenches

The cliche in football is that games are won in the trenches. For the Titans offensive line, they were able to generate positive EPA/rush on the left side of their line with Skoronski and Latham which led to a 26 yard rushing touchdown by Pollard off the left guard. The right side of the line was plagued by injury as starting right guard Dillon Radunz was only able to play 76% of the snaps. Continuity on the offensive line is imperative in its performance, so hopefully Radunz will not have to miss any snaps because of this injury. I would expect the Titans to try to run behind Latham in this game and stay away from 2-time Pro-Bowler Quinnen Williams.

The Jets were able to generate success behind veteran tackle Tyron Smith; however, there weren't many other bright spots on their offensive line. I would expect T'vondre Sweat to have a big game against this interior offensive line which ranked 24th on inside runs last week while the Titans ranked 4th in defending interior runs.

# Passing Heatmap

Levis did not have a great day. He was 19/32 for only 127 yards and 1 touchdown with 2 interceptions. The low yardage was caused by an average depth of target of 7 yards compared to his league-leading 10 yards per target last season. As mentioned above, Levis had a large percentage of his yards from play action last season and it might be worth utilizing play action more this week to open up more passing options for Levis. The Jets defense struggled on non-Sauce targets with below-average EPA in these situations. I would expect the Titans to throw away from Sauce early in the game to build Levis' confidence and get some positive EPA short throws.

The Titans pass defense was unable to generate any interceptions last week; however, they did not allow a completion over 11 yards and held Caleb Williams expected completion percent to 64% which was ranked 20th. This week, they face future Hall of Famer Aaron Rodgers who had an expected completion percent of 72.6% which ranked 7th. Rodgers refused to throw to the right-side of the field against the 49ers in week 1; however, the Titans should expect favorable matchups on both sides of the field as they will have Sneed or Chido on any non-Garrett Wilson pass catcher. 

# Pass Catchers

Chig was the only pass catcher to bring in a receiving touchdown in week 1 while not seeing much volume. Ridley led the way with 7 targets and the backfield had a healthy share of targets as well with 8 combined betwen Pollard and Spears. With Hopkins still working his way back from an injury, I would expect his targets to increase which will help relieve some of the pressure from Ridley. 

For the Jets, their targets were centered around Lazard, Wilson, and Hall. I would expect more of the same against the Titans this week because Mike Williams is still working his way back from an injury and 2024 draft pick Malachi Corely will only be in his second professional game. While Breece Hall had a negative EPA/catch last week, it is difficult to determine how successful he will be this week because the Titans only faced 2 running back targets last week which led to 0.09 EPA/catch.

# Rushers

The Titans talked this off season about splitting reps with their running backs and that is what we saw with Pollard taking 60% of the snaps and Spears playing 45% of the snaps. The overlap is caused by 21 personnel which saw Pollard split out to receiver for some snaps. While the snap counts were evenly split, it was Pollard who took the larger share of rushes last week as he played more of the early downs than Spears. I would like to see more of the 21 personnel look and use Pollard in a similar role to Christian McCaffrey where he can share the backfield with Spears as well as line up as a receiver. I think the 49ers are able to get defenses into packages to cover 21 personnel which might not work if the final formation resembles 11 personnel.

For the Jets, they opted for a workhorse backfield playing Breece Hall over 85% of the snaps and taking all but 3 carries. Unfortunately for Hall, he was not able to get much going on the ground. I would expect Hall to have another slow day on the ground as the Titans were able to hold D'Andre Swift to -0.42 EPA/rush on 10 carries.

# Prediction

This game will be determined on the defensive side of the football. The Jets offense struggled last year and was mediocre last week; therefore, I think the new-look Titans defense continues its dominance in week 2. On the other side, the Jets defense was one of the better defenses in the league last season and was able to add Haason Reddick in the offseason while maintaining much of their production. I would expect some regression to that this week as they get a much easier offense to face. I think the Titans cover +3.5 in this game with the total well below 40.5.
