# Week 4 vs Miami

Week 4 is here and the season is in full swing. Fantasy managers are in a rythym of getting their waiver pickups in every week before Wednesday morning. Survival league participants are also in a ryhtym of getting their weekly picks selected, if they're lucky enough to have made it this far. As for the Titans, they are still seeking their first win. Fortunately, they travel to Miami to face a depleted Dolphins team who could potentially have to turn to their third starting QB of the season because of injury. 

# Overall Preview

The Titans countinue to have woes in the passing game as they now rank last in EPA/pass. This has been as the result of the points lost via interception; the Titans have lost about 20 expected points via interception which puts them in second to last in EPA lost via interception. However, this isn't bad turnover luck. The Titans rank 24th in interception-worthy throws as charted by FTN. After week 1, I suggested that the Titans use more play action to free up the receivers for Levis; however, after a few weeks into the season, the Titans are using play action at the 17th highest rate per pass. This usage has not led to success; rather, the Titans have the 30th worst EPA/play action pass. This could be as a result of defenses not blitzing much. The Titans face the 7th fewest pass rushers per pass which allows defenses to drop more into coverage. This low blitz rate could be as a result of the Titans having the 13th highest pass rate over expected. Such a high pass rate would cause the defenses to put fewer players in the box resulting in fewer pass rushers. Going into this matchup, I think the Titans should utilize the run more. Rushing is the less efficient option on average; however, the Titans have been able to generate the 14th highest EPA/rush againts 7-man boxes and 18th highest EPA/rush against 6-man boxes. More success against the box numbers would cause defenses to bring more into the box.

On defense, the Titans look to get back on track after allowing 23 points and now ranking 16th in EPA/play allowed. While the offense has seemed to stall in the second half, the defense has still only been on the field for the 14th highest amount of time. Thus, fatigue might not be playing a large role in the defense's recession. On another note, the Dolphins could potentially have to bring in their third starting quarterback of the season as a result of injuries. The Titans have been successful against the pass, and I would expect theses factors to cause the Dolphins to take a run-first approach. While the Titans have not been as successful against the run as they are against the pass, compared to the rest of the league, the Dolphins rush offense has been even worse.

# Trenches

Coming into the season, it was expected that much of the Titans run success would come from the left side of the line as that is where most of the draft captial has gone in the last two drafts. Therefore, it is no surpise that the Titans are most successful when running behind left tackle and left guard compared to the rest of the league in these gaps. It will be interesting to see how the Dolphins play this as they have had the most success on the opposite side of the line. Will they put their best againts the Titans' best, or will they pressure the right side of the Titans offensive line that has yet to see much success this season?

The Dolphins have not had much success running the ball this season. As a result of having to run the ball more, I would expect the Dolphins to take the pressure off their running backs and bring in their fast receivers to receive sweeps. This also might be a decent plan for the Dolphins as the Titans have been quite successful at defending the inside run because of T'Vondre Sweat.

# Passing Heatmap

Will Levis has taken some time to get going this season, but he was able to have an average game against Green Bay. Levis finished with the 18th best success rate in week 3; however, interceptions and sacks caused Levis to only have the 27th ranked EPA/pass as he lost 8 expected points on his at the Titans' 31 yardline in the 1st quarter. I would expect the Titans to target the intermediate areas against the Dolphins because the Titans rank 8th in EPA/pass when the air yards are between 5 and 20 yards whereas the Dolphins rank 26th in EPA/pass allowed in this same range.

As for the Dolphins passing offense, they have experienced success throwing to Jaylen Waddle on the right side of the field. The Titans have been stronger on the left side of the defense which could cause issues for the Dolphins with a new quarterback. Unfortunately, the Titans' pass rush has not been able to help out their pass defense. Their pass rush is ranked 22nd in pressure rate while ranking 11th in blitz percent per dropback.

# Pass Catchers

Most of the efficiency for the Titans' pass catchers has come from the tight ends as 4 of the Titans top 5 most efficient pass catchers are tight ends. The most utilized pass catcher has been Calvin Ridley; however, the Titans have distributed the ball fairly evenly among their wide receivers. While DeAndre Hopkins was underutilized in the first two weeks, he received the majority of targets in week 3 leading the way with 7 targets.

The Dolphins have utilized a plethora of pass catchers throughout the first 3 weeks of the season, although some have received few targets. Jaylen Waddle has been the most efficient pass catcher among those with significant targets. The pass catchers were able to be decently efficient with Tua at quarterback; however, last week with a backup, they were not as efficient with only two receivers having a positive EPA/catch. This could be beneficial to Titans secondary that has been about leage average.

# Rushers

Tony Pollard has received the majority share of rushes for the Titans. It seems that while Pollard and Spears are splitting snaps, Pollard will be who the Titans lean toward for rushing. Will Levis has been an efficient rusher on designed run plays; however, sacks have been an issue for Levis as he ranks 35th among all quarterbacks in EPA lost via sack.

The Dolphins have been going with DeVon Achane for running. This has not been a successful approach as Achane ranks 38th out of 58 rushers with more than 10 carries which could be the reason why the Dolphins have tried to bring others into their running game. 

# Prediction

I predict the Titans to cover +2.5 this week and the game to go over 36.5. I think the Dolphins will be predictable this week, whether that's with Skylar Thompson or Tim Boyle, which puts an already struggling rush offense at a disadvantage. It could also allow the Titans' weak pass rush to bring more pass rushers. Levis also showed promise in week 3, and he faces another weak pass defense which could allow him to continue to improve. I think the Dolphins weak pass rush will also allow Levis to reduce his sack numbers which have hindered his performance in past weeks.
